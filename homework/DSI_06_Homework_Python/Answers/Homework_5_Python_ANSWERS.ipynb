*We can use the function train_test_split() to split the observations into two halves for the training and testing sets.*

```python
# Set a random seed for reproducibility
np.random.seed(16)

# Split the observations into training and test sets
total_rows = len(College)
train_rows = total_rows // 2  # Integer division to get roughly half of the rows

# Generate a random array of indices for the training set
train_indices = np.random.choice(total_rows, train_rows, replace=False)

# Create the test set by excluding the indices used for the training set
test_indices = np.setdiff1d(np.arange(total_rows), train_indices)

# Create the training and test sets using iloc
train_set = College.iloc[train_indices]
test_set = College.iloc[test_indices]

# Print the lengths of the training and test sets
print("Number of observations in the training set:", len(train_set))
print("Number of observations in the test set:", len(test_set))
```
