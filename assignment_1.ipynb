### Question 1: Simple Linear Regression 

Let's set up our workspace and use the `Boston` dataset in the `ISLP` library. Print `Boston` to learn more about the dataset.

```python
# Import standard libraries
import numpy as np
import pandas as pd
import seaborn as sns
from matplotlib.pyplot import subplots
import statsmodels.api as sm
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression

# Import specific objects
from sklearn.preprocessing import StandardScaler
from sklearn.neighbors import KNeighborsClassifier

from statsmodels.stats.outliers_influence \
     import variance_inflation_factor as VIF
from statsmodels.stats.anova import anova_lm

from ISLP import load_data
from ISLP.models import (ModelSpec as MS,
                         summarize,
                         poly)
```

```python
# Load the "Boston" dataset using the "load_data" function from the ISLP package
Boston = load_data('Boston')
```

Before we fit and review model outputs, we should visualize our data. Review the code and plot, shown below. Answer the following questions:

_(i)_ What are the `medv` and `dis` variables being plotted? (Hint: review this [link](https://islp.readthedocs.io/en/latest/datasets/Boston.html))

medv: median value of owner-occupied homes in $1000s.
dis: weighted mean of distances to five Boston employment centres.


_(ii)_ What concept ‘defines’ the plotted line?
The plotted line represents the line of best fit according to the least square criterion. 

```python
# Extract the variables
medv = Boston['medv'].values.reshape(-1, 1)
dis = Boston['dis'].values.reshape(-1, 1)

# Plot data
plt.scatter(dis, medv, label='Data')
plt.xlabel('dis')
plt.ylabel('medv')

# Fit a linear regression model
lm = LinearRegression()
lm.fit(dis, medv)

# Plot the regression line
plt.plot(dis, lm.predict(dis), color='red', label='Regression Line')

# Add labels and legend
plt.xlabel('dis')
plt.ylabel('medv')
plt.legend()

# Show the plot
plt.show()
```

Consider the variables plotted above. In the context of the `Boston` dataset:

_(iii)_ What is the (implied) null hypothesis? What is the (implied) alternative hypothesis?

-The null hypothesis is that there is no significant relationship between the weighted mean of distances to five Boston employment centres and the median value of owner-occupied homes in $1000s. 

-The alternative hypothesis is that there is a significant relationship between the weighted mean of distances to five Boston employment centres and the median value of owner-occupied homes in $1000s. 



_(iv)_ Now, let’s fit a simple regression model, using the general syntax `sm.OLS()` and `model.fit()`. As above, use `medv` as the response variable Y, and `dis` as the predictor variable X. (Hint: use the syntax `sm.add_constant` with the appropriate argument).

```python
# Add your code here

medv = Boston['medv'].values.reshape(-1, 1)
dis = Boston['dis'].values.reshape(-1, 1)

#Add an intercept to the predictor
dis_with_constant = sm.add_constant(dis)

#Fit the linear reg model
model = sm.OLS(medv, dis_with_constant)
results = model.fit()

print(results.summary())
```

### Question 2: Multiple Linear Regression 

We'll continue to use the `Boston` dataset for this question.

_(i)_ Fit a multiple linear regression, with two predictor variables: $X_1$ is `dis`, and $X_2$ is `rm`. As before, keep `medv` as the response variable Y. (Hint: use the syntax `sm.add_constant` with the appropriate arguments).

```python
# Add your code here

#variables
medv = Boston['medv'].values.reshape(-1, 1)
dis = Boston['dis'].values.reshape(-1, 1)
rm = Boston['rm'].values.reshape(-1,1)

#create a df and add a constant
X = Boston[['dis','rm']]
X = sm.add_constant(X)

#Fit the multiple linear reg model
MLR = sm.OLS(medv,X).fit()

#print results
print(MLR.summary())
```

_(ii)_ In the context of the `Boston` dataset, state the null and alternative hypotheses.

The null hypothesis is that there is no significant relationship between dis(distance to 5 employment centres), rm (avg no. of rooms per dwelling) and medv (median value of owner-occupied homes).
The alternative hypothesis is that there is a significant relationship between dis, rm and medv.

_(iii)_ Review the model output, using `summary()`. Does it appear that both `dis` and `rm` are predictive of `medv`? How did you determine this?

Yes. Since the p-values of both 'dis' and 'rm' are 0.001 and 0.000 respectively, both are less than 0.05. This indicates that we can reject the null hypothesis for both predictors. 
The coefficient of 'dis' is 0.4888, it indicates that as the weighted mean of distance to employment centres increases, the median value of homes increases, holding 'rm' constant.
The coefficient of 'rm' is 8.8014, it indicates that as the number of room per dwelling increases, the median value of homes increases, holding 'dis' constant.

_(iv)_ We can use the inbuilt `sm.graphics.plot_regress_exog` function to generate helpful diagnostic plots (Hint: provide `plot_regress_exog` with the multiple regression model). Review the first generated plot, 'Residuals vs. Fitted'. Which observations are outliers? What impact might outliers have on our model?

In the first plot, we can see that there are a few outliers with 'medv' of 45 to 50. They deviate substantially from the overall pattern that clusters between 20 to 30 medv. 
Outliers can affect the model complexity. If the model tries to fit the outliers, it will lead to overfitting. If the model ignores them, it will lead to underfitting.
Outliers can also affect the model fit by disporportionately affecting the relatoinship between the predictors and response variable, resulting in poor fit.

```python
# Add your code here

#Plot
import statsmodels.api as sm
from statsmodels.graphics.regressionplots import plot_regress_exog
import matplotlib.pyplot as plt

fig = plt.figure(figsize=(12,8))
plot_regress_exog(MLR, 'dis', fig=fig)
plt.show()
```

_(v)_ Fit a second model, this time including an interaction between the two predictor variables. Is there an interaction? (Hint: add a variable `x1 * x2` where `x1` and `x2` are the predictor variables). State an interpretation of the interaction, in the context of the `Boston` dataset, in one or two sentences.

Yes, there is an interaction between 'dis' and 'rm'. The p-value of dis_rm_interaction is less than 0.05, suggesting that the interaction term contributes significantly to explaining the variability of 'medv'.

The coefficient of the interaction dis_rm_interaction is 0.9855, with a standard error of 0.212.
This shows that for each unit increase in 'dis'*'rm', the median value of homes increases by approximately 0.9855, holding 'dis' and 'rm' constant.

```python
# Add your code here

#Create the interaction variable
Boston['dis_rm_interaction'] = Boston['dis'] * Boston['rm']

#add constant
X_interaction = sm.add_constant(Boston[['dis', 'rm', 'dis_rm_interaction']])

#fit the model
MLR_interaction = sm.OLS(medv, X_interaction).fit()

#print the results
print(MLR_interaction.summary())
```
